```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random
import numpy as np

data = {
    'Сайт': [f"Сайт {random.randint(1, 5)}" for _ in range(100)],
    'Куратор': [f"Куратор {random.randint(1, 5)}" for _ in range(100)],
    'Тип отзыва': [random.choice(['Положительный', 'Отрицательный']) for _ in range(100)],
    'Подведомственность': [random.choice(['Подведомственность A', 'Подведомственность B']) for _ in range(100)],
    'Уровень': [random.choice(['Уровень A', 'Уровень B']) for _ in range(100)],
    'Дата отзыва': pd.date_range(start='2023-01-01', end='2023-01-10', periods=100),
    'Ответ куратора': [random.choice([None, f"Ответ {i}"]) for i in range(100)]
}

df = pd.DataFrame(data)
```

```python
top5_sites = df.groupby('Сайт').size().sort_values(ascending=False).head(5)
print("ТОП 5 сайтов по числу отзывов:")
print(top5_sites)
```

```python
top5_curators = df.groupby('Куратор').size().sort_values(ascending=False).head(5)
print("\nТОП 5 кураторов по числу отзывов:")
print(top5_curators)
```

```python
df['нет ответа'] = df['Ответ куратора'].isna()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

sns.barplot(x="Тип отзыва", y="нет ответа", hue="Подведомственность", data=df)
plt.title("Число отзывов без ответа с разбивкой по типу отзыва и подведомственности")
plt.show()
```

```python
df['Дата отзыва'] = pd.to_datetime(df['Дата отзыва'])
```

```python
df['День недели отзыва'] = df['Дата отзыва'].dt.dayofweek
```

```python
df.groupby('День недели отзыва').size().plot(kind='bar')
plt.xticks(range(7), ['Пн', 'Вт', 'Ср', 'Чт', 'Пт', 'Сб', 'Вс'])
plt.title("Число отзывов по дням недели")
plt.xlabel("День недели")
plt.ylabel("Число отзывов")
plt.show()
```

```python
pivot_table = df.pivot_table(index='День недели отзыва', columns='Тип отзыва', values='Сайт', aggfunc='count')
pivot_table.plot(kind='bar', stacked=True)
plt.xticks(range(7), ['Пн', 'Вт', 'Ср', 'Чт', 'Пт', 'Сб', 'Вс'])
plt.title("Распределение числа отзывов по дням недели с разбивкой по типу отзыва")
plt.xlabel("День недели")
plt.ylabel("Число отзывов")
plt.legend(title="Тип отзыва")
plt.show()
```
